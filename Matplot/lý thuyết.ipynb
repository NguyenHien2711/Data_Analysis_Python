```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
students = pd.read_csv('students.csv')
students.head(1)
```

### 3. Mảng đồ thị

#### Nhiều biểu đồ cùng loại trên cùng một Axes

#### Nhiều biểu đồ cùng loại trên 1 figure
Nghĩa là nhiều Axes trên 1 figure

```python
fig, ax = plt.subplots(2,1) #2: hàng 2 axes, 1 cột
ax[0].plot(students.index, students['Toán'], color='r', label='Toán')
ax[0].plot(students.index, students['Lý'], color='b', label='Lý')
ax[0].plot(students.index, students['Hoá'], color='g', label='Hoá')
ax[0].set_ylabel('Khối A', fontsize=14)
ax[0].legend(fontsize=14, loc='lower right')

ax[1].plot(students.index, students['Văn'], color='r', label='Văn')
ax[1].plot(students.index, students['Sử'], color='b', label='Sử')
ax[1].plot(students.index, students['Địa'], color='g', label='Địa')
ax[1].set_ylabel('Khối C', fontsize=14)
ax[1].legend(fontsize=14, loc='lower right')

fig.set_size_inches(18.5, 6.5)
plt.show()
```

#### Nhiều biểu đồ khác loại trên 1 Axes

```python
students['Điểm Trung Bình']= ((students['Toán'] + students['Văn'] + students['Tiếng Anh'])*2 +students['Lý']+students['Hoá'] + students['Sinh'] + students['Sử']+ students['Địa'])/11
students.head(1)
```

```python
#dựng 1 bar chart bth
plt.bar(students.index, students['Điểm Trung Bình'], label='Điểm TB', width=0.5)
# lấy 1 axes, gca, get current axes
axes1 = plt.gca()
#Duplicate axes2 bằng twiny=> Để mục đích nằm trên axes1 luôn
axes2 = axes1.twiny()
axes2.plot(students.index, students['Toán'], label = 'Điểm Toán', linewidth=2, color='r',marker='o')
plt.legend(loc='upper right')
axes1.set_ylabel('Thang điểm 10', fontsize=10)
plt.show()
#tHƯỜNG có trục X chung nhau
```

#### Nhiều biểu đồ khác loại trên 1 Figure

```python
fig, ax =plt.subplots(1,2)
ax[0].bar(students.index, students['Điểm Trung Bình'], label='Điểm TB', width = 0.5)
ax[0].set_ylabel('Thang điểm 10')
ax[0].legend(loc='upper right')
ax[0].set_title('Điểm TB')
ax[0].tick_params(axis='both', size=14)

gioi = students[students['Toán']>=8]['Mã học sinh'].count()
kha = students[(students['Toán']<=8.5)&(students['Toán']>=6.5)]['Mã học sinh'].count()
tb = students[(students['Toán']<=6.5)&(students['Toán']>=5)]['Mã học sinh'].count()
kem = students[students['Toán']<5]['Mã học sinh'].count()
#Đặt tên cột dính gần nhau dễ sửa hơn
scores=[gioi, kha, tb, kem]
labels= ['Giỏi', 'Khá', 'Trung bình', 'Kém']

ax[1].pie(scores, labels=labels, )
fig.set_size_inches(18.5, 6.5)
plt.show()
```
