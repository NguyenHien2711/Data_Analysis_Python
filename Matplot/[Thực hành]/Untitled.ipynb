```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('daily-min-temperatures.csv')
data
```

```python
plt.hist(data['Temp'], bins =45, range=(1,23), histtype = 'step')
plt.title('Temperature Histogram', fontsize=16)
plt.show()
```

The value that appears most often is in the range of 10-11 degrees.<br>
The chart show that there are some abnormal value in both side, some days have extremely more higher temperature above 23 degrees and lower under 1 degree<br>
Overal, if outliers are extracted, it can be said that temperature follow a standard normal distribution<br>

```python
plt.plot(data['Date'], data['Temp'])
plt.title('Daily Temperature', fontsize =16)
plt.xlabel('Date', fontsize = 14)
plt.xlabel('Temp', fontsize = 14)
plt.show()
```

As we can see, changes in the temperature have followed a cyclical pattern.

```python
data['Month'] = pd.to_datetime(data['Date'], format = '%Y-%m-%d').dt.month
data['Year'] = pd.to_datetime(data['Date'], format = '%Y-%m-%d').dt.year
d1 = data[data['Year'] == 1990]
d1 = d1.groupby('Month')['Temp'].mean()
# d1 = data[(data['Date'] >= bounds[0]) & (data['Date'] <= bounds[1])]
plt.plot(d1.index, d1.values)
plt.title('Daily Temperature in 1990', fontsize = 16)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Temp', fontsize = 14)
plt.show()
```

Daily Temperature tend to considerablly dip from May and start to increase again in September
Daily Average Temperature have the highest degrees during 3 first months of the year.
