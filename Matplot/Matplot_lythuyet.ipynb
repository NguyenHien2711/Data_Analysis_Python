```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
plt.plot(x, np.sin(x)) #biểu đồ đường
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), label='sin(x)') # label để gọi legend thì sẽ show ra 
plt.plot(x, np.sin(x), label='cos(x)')
plt.title('Hàm sin(x), cos(x)...', fontdict={'fontsize':12, 'color':'blue'})
plt.xlabel('x values')
plt.ylabel('sin(x), cos(x) values')
plt.legend()
plt.show()
#end show rồi nếu làm cell khác không liên quan
```

### Thay đổi cài đặt Axes, Axis

```python
#get curcent axes
#TH có nhiều axes thì phải get dùng hàm gca()
#Đang chỉnh sửa khung
# plt.plot(x, np.sin(x), label='sin(x)') # label để gọi legend thì sẽ show ra 
# plt.plot(x, np.sin(x), label='cos(x)')
axes = plt.gca()
axes.set_xlabel('Giá trị đầu vào', color='red', fontsize=14) # label cua axes
axes.set_ylabel('Giá trị đầu vào', color='green', fontsize=14)
axes.tick_params(axis='x', colors='blue', size =12) # hiện ở chỗ axis
axes.tick_params(axis='y', colors='black', size =12)
plt.show()
```

### Adjusting line colors and style

```python
plt.plot(x, np.sin(x-0), color='blue') #specify by name nhiều
plt.plot(x, np.sin(x-1), color='g') # Short color code nhiều
plt.plot(x, np.sin(x-2), color='0.75') #Graysale between 0 and 1
plt.plot(x, np.sin(x-3), color='#a150ec') 
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
plt.plot(x, x+1, linestyle='solid') #mặt định
plt.plot(x, x+2, linestyle='dashed')
plt.plot(x, x+3, linestyle='dashdot')
plt.plot(x, x+4, linestyle='dotted')
```

```python
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+4, linestyle='-')
```
