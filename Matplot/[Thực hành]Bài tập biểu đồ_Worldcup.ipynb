```python
#Vẽ biểu đồ so sánh tỉ lệ giữa đội giành thấng 
# trong hiệp ` và tỉ lệ đội giành chiến thắng cuối 
# mỗi trân
data = pd.read_csv('WorldCupMatches.csv')
data.head()
data.drop_duplicates(inplace=True)
data.head()
```

```python
data['Home Team Name'].replace(to_replace='Germany FR', value='Germany', inplace=True)
data['Away Team Name'].replace(to_replace='Germany FR', value='Germany', inplace=True)
```

```python
halftime_win_home_team = data[data['Half-time Home Goals']>data['Half-time Away Goals']]['Home Team Name']
halftime_win_away_team = data[data['Half-time Home Goals']<data['Half-time Away Goals']]['Away Team Name']
halftime_win = pd.concat([halftime_win_home_team, halftime_win_away_team], axis=0)
halftime_win.value_counts()
```

```python
halftime_win_freq = halftime_win.groupby(lambda x: halftime_win[x]).count()
#x là chỉ số của series vì thường group by của series mặc định sẽ groupby theo x
halftime_win_freq.sort_values(inplace=True, ascending=False)
halftime_win_freq
```

```python
data['result'] = data['Home Team Goals']-data['Away Team Goals']
condition = [data['result']>0, data['result']<0, data['result']==0]
choices = [data['Home Team Name'], data['Away Team Name'],data['Win conditions'].str.split(' ').str[0]]
data['Winfinal_country']=np.select(condition, choices)
data
```

```python
fulltime_win_home_team = data[(data['Half-time Home Goals']>data['Half-time Away Goals'])
                             &(data['Home Team Goals']>data['Away Team Goals'])]['Home Team Name']
fulltime_win_home_team
```

```python
fulltime_win_away_team = data[(data['Half-time Home Goals']<data['Half-time Away Goals'])
                             &(data['Home Team Goals']<data['Away Team Goals'])]['Away Team Name']
fulltime_win_away_team
```

```python
fulltime_win_team= pd.concat([fulltime_win_home_team, fulltime_win_away_team], axis=0)
fulltime_win_team_freq=fulltime_win_team.value_counts()
```

```python
fig,ax = plt.subplots(1,2)
fig.set_size_inches(18, 6.5)
ax[0].bar(halftime_win_freq[:5].index, halftime_win_freq[:5].values,label='Halftime-win')
ax[0].bar(fulltime_win_team_freq[:5].index, fulltime_win_team_freq[:5].values,label='Fulltime-win')
ax[0].set_ylabel('Số lần thắng hiệp 1, cả trận')
ax[0].legend()
#Ghi giá trị của mỗi ba 
#thuọc tính container là chiều cao của các cột, bar là gtr chiều cao của từng cột
#bar_label là gtr ghi trên cột 
for bars in ax[0].containers:
    ax[0].bar_label(bars)
ax[1].bar(x=halftime_win_freq[:5].index, y=fulltime_win_team_freq[:5].values/halftime_win_freq[:5].values)
ax[1].set_ylabel('Phần trăm')
for bars in ax[1].containers:
    ax[1].bar_label(bars)

plt.show()
```
